## 9.6

```python
import re

s = 'English = 89, Science = 90, Math = 92, History = 80.'

score_str = re.findall('[0-9]+', s) #숫자 문자열 리스트 생성
score_int = [] #숫자 정수자료형 리스트 생성

for i in score_str :
    score_int.append(int(i)) #정수로 만들어서 리스트에 넣는다.

print('문장 s :', s)
print('총점 :', sum(score_int)) #리스트 속 숫자의 총합 
print('평균 점수 : ', sum(score_int) / len(score_int))
```

## 9.7

Veni, vidi, vici.

```python
import string

src = input("문장을 입력하시오 : ")
b = int(input("이동시킬 칸 수를 입력하시오 : ")) # 이동시킬 칸 수도 입력 받음

src_str = string.ascii_uppercase + string.ascii_lowercase # 대문자 소문자 둘 다 가능하게 수정
dst_str = src_str[b:] + src_str[:b] # 이동시킬 칸 수 넣어줌

def cipher(a):
    idx = src_str.index(a)
    return dst_str[idx]

print("암호화된 문장 :", end='')

for ch in src:
    if ch in src_str:
        print(cipher(ch), end='')
    else:
        print(ch, end='')
        
print()
```

## 9.8

```python
import re

s = input("문장을 입력하시오 : ")

a = re.findall('[A-z]+\S\s', s) # 영문 단어
b = re.findall('\s\d+', s) # 숫자
c = re.findall('\S[A-z]+\d+', s) # 영문자 + 숫자

print('영문 단어 : ', end = '')
for i in a :
    print(i, end = '')

print("")
print('숫자 :', end = '')
for i in b :
    print(i, end = '')

print('')
print('영문자 + 숫자 : ', end = '')
for i in c :
    print(i, end = '')
```

## 9.9

```python
import wikipedia

# 위키백과 사전의 컨텐츠 제목을 명시해 준다
wiki = wikipedia.page('South Korea')

# 이 페이지의 텍스트 컨텐츠를 추출하도록 한다
text = wiki.content
```

```python
from wordcloud import WordCloud, STOPWORDS

# 중지어가 제외된 워드 클라우드를 만들자
s_words = STOPWORDS.union({'south', 'north', 'korean', 'world', 'country'})
wordcloud = WordCloud(width = 2000, height = 1500, stopwords = s_words).generate(text)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(40, 30))

# 화면에 이미지를 그려준다
plt.imshow(wordcloud)
plt.show()
```

## 9.10

힌트 : 두 개의 기사 페이지를 병합하면 된다

```python
import wikipedia

# 위키백과 사전의 컨텐츠 제목을 명시해 준다
wiki1 = wikipedia.page('South Korea')
wiki2 = wikipedia.page('Japan')

# 이 페이지의 텍스트 컨텐츠를 추출하도록 한다
text = wiki1.content + wiki2.content
```

```python
from wordcloud import WordCloud, STOPWORDS

# 중지어가 제외된 워드 클라우드를 만들자
s_words = STOPWORDS.union({'south', 'north', 'korean', 'world', 'country'})
wordcloud = WordCloud(width = 2000, height = 1500, stopwords = s_words).generate(text)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(40, 30))

# 화면에 이미지를 그려준다
plt.imshow(wordcloud)
plt.show()
```

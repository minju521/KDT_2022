## 11.2

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0, (2 * np.pi) * 6 + 2, 0.1) #6번 반복 / 0.1 간격으로 x 생성
plt.plot(a, a * np.sin(a), color = 'red') # a가 증가하므로 sin(a)에 a를 곱하면 증가한다.
plt.plot(a, 20 * np.cos(a), '--', color = 'blue') #진폭 20 
plt.show()
```

```python
# subplots 이용하여 그래프 3개 한번에 그리기
import numpy as np
import matplotlib.pyplot as plt

x1 = 6 * np.random.randn(1000) + 25
y1 = 6 * np.random.randn(1000) + 25

x2 = 6 * np.random.randn(1000) + 25
y2 = 3 * np.random.randn(1000) + 25

x3 = 6 * np.random.randn(1000) + 25
y3 = x3 + np.random.randn(1000)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].scatter(x1, y1, color = 'blue')
axs[1].scatter(x2, y2, color = 'red')
axs[2].scatter(x3, y3, color = 'green')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
 
mu = 25          #평균은 전부 25이다. 
sigma1 = 6       #표준편차1 
sigma2 = 3       #표준편차2 
 
#정규분포 = 평균 + 표준편차 * 수 
x = [mu + sigma1 * np.random.randn(1000)]
y1 = [mu + sigma1 * np.random.randn(1000)]
y2 = [mu + sigma2 * np.random.randn(1000)]
y3 = [x + np.random.rand(1000)]
 
#서브플롯으로 나타낸다. (행이 1, 열이 3) 
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
 
ax[0].scatter(x, y1)
ax[1].scatter(x, y2, color = 'red')
ax[2].scatter(x, y3, color = 'green')
plt.show()
```

## 11.5

```python
# 개 종류와 크기 데이터 입력(몸통 길이와 높이)

# 닥스훈트
dach_length = [77, 78, 85, 83, 73, 77, 73, 80]
dach_height = [25, 28, 19, 30, 21, 22, 17, 35]

# 사모예드
samo_length = [75, 77, 86, 86, 79, 83, 83, 88]
samo_height = [56, 57, 50, 53, 60, 53, 49, 61]

# 말티즈
malt_length = [34, 38, 38, 41, 30, 37, 41, 35]
malt_height = [22, 25, 19, 30, 21, 24, 28, 18]
```

(1)

```python
import matplotlib.pyplot as plt

plt.scatter(dach_length, dach_height, color='r', marker='o')
plt.title("Dachshund size")
plt.xlabel("Length")
plt.ylabel("Height")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(samo_length, samo_height, color='b', marker='^')
plt.title("Samoyed size")
plt.xlabel("Length")
plt.ylabel("Height")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(malt_length, malt_height, color='g', marker='s')
plt.title("Maltese size")
plt.xlabel("Length")
plt.ylabel("Height")
plt.show()
```

(2)

```python
import matplotlib.pyplot as plt

plt.scatter(dach_length, dach_height, color='r', marker='o', label='Dachshund')
plt.scatter(samo_length, samo_height, color='b', marker='^', label='Samoyed')
plt.scatter(malt_length, malt_height, color='g', marker='s', label='Maltese')

plt.title("Dog size")
plt.xlabel("Length")
plt.ylabel("Height")

plt.legend()
plt.show()
```

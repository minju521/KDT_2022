## 14.1 이세돌을 이긴 컴퓨터 프로그램 : 비결이 뭘까

컴퓨터가 데이터를 기반으로 스스로 학습할 수 있다면 컴퓨터는 더욱더 복잡한 일을 할 수 있을 것이다.

기계학습은 인공지능의 한 분야로, 컴퓨터에 학습 기능을 부여하기 위한 연구 분야이다. 학습에 사용할 수 있는 데이터가 많아지면 판단을 내리는 알고리즘 성능이 향상될 것이다.

학습 알고리즘은 항상 정해진 동작을 수행하는 명령어로 구성된 알고리즘과 달리 데이터를 이용하여 예측하고 판단할 수 있게 된다.

## 14.2 다양한 종류의 기계학습 기법

기계학습은 일반적으로 가르쳐주는 '교사'의 존재 여부에 따라 크게 지도학습과 자율학습으로 나누어진다. 그리고 보상과 처벌에 따른 강화학습으로 분류할 수 있다.

- 지도학습 : 컴퓨터는 '교사'에 의해 주어진 예제와 정답을 제공받는다. 지도학습의 목표는 입력을 출력에 매핑하는 일반적인 규칙을 학습하는 것이다.
- 자율학습 : 외부에서 정답이 주어지지 않고 학습 알고리즘이 스스로 입력에서 어떤 구조를 발견하는 학습니다. 흔히 비지도학습이라 부른다. 자율 학습의 대표적인 예는 클러스터링이다.
- 강화학습 : 보상 및 처벌의 형태로 학습 데이터가 주어진다.

## 14.3 지도 학습은 문제와 정답을 알려주고 학습을 시키자

지도학습은 주어진 입력-출력 쌍을 학습한 후에 새로운 입력값이 들어왔을 때, 합리적인 출력값을 예측하는 것이다. 즉 지도 학습은 입력과 출력이 주어질 때, 입력에서 출력으로의 매핑 함수를 학습하는 것이라 할 수 있다.

## 14.4 데이터를 잘 설명하는 함수를 찾아라 : 회귀문제

지도 학습 알고리즘에는 회귀 분석과 분류의 두 가지 유형이 있다. 회귀는 연속적인 값을 예측하고, 분류는 객체를 어떤 카테고리 중의 하나로 예측한다.

회귀란 일반적으로 데이터들을 다차원 공간에 표시한 후에 데이터들을 가장 잘 설명하는 직선이나 곡선을 찾는 문제이다. 입력 x와 출력 y를 보면서 함수를 예측하는 것을 회귀 기법이라고 한다.

선형회귀는 기계학습으로 생각하기에는 너무 단순하고 단순히 통계적 방법이라고 생각하는 사람들도 있다. 하지만 회귀 문제도 입력x에 대응하는 실수 y가 주어지고 함수를 학습하는 것이므로 일종의 기계학습 문제로 생각할 수 있다.

데이터를 학습시킬 때, 데이터를 원형 그대로 사용하는 경우도 있지만 일반적으로는 데이터에서 어떤 특징을 추출하여 이것으로 학습시키고 테스트하게 된다. 특징이란 관찰되는 현상에서 측정할 수 있는 개별적인 속성을 의미한다.

## 14.5 가장 간단한 회귀 : 선형 회귀 분석

사이킷런은 파이썬 언어에서 기계학습을 수행하는 라이브러리이다. 이 라이브러리는 K-NN 알고리즘, 서포트 백터머신, 랜덤 포레스트, K-means 등의 분류, 회귀, 클러스터링 알고리즘을 포함하고 있어서 기계학습을 처음 접하는 사람들에게 좋은 도구이다.

선형회귀는 임의의 변수 x와 이 변수에 따른 또 다른 변수 y와의 상관관계를 모델링하는 기법이다. 두 변수의 관계를 알아내거나 이를 이용하여 y가 없는 x값에 대해 y를 예측하는데 사용할 수 있다.

y = mx + b 에서 m은 직선의 기울기이고 입력 변수에 곱해지는 계수이다. b는 절편이다. 기본적으로 선형 회귀 알고리즘은 데이터를 설명하는 가장 적절한 기울기와 절편값을 찾는다. 기본적으로 선형 회귀 알고리즘은 데이터 요소에 여러 직선을 맞추어 본 후에 가장 적은 오류를 발생시키는 직선을 반환한다.

2개 이상의 변수가 있는 경우까지 확장될 수 있는데 이를 다중회귀분석이라고 한다. 2차원 공간에서 선형 회귀 모형은 직선이고 3차원에서는 평면이고, 3차원 이상에서는 초평면이다.

## 14.6 선형 회귀를 사이킷런 라이브러리로 구현해 보자

사이킷런을 코드에 가져오기 위해서는 sklearn이라는 이름으로 가져와야 한다. 선형회귀를 위해 가장 먼저 해야 할 작업은 사이킷런 라이브러리와 넘파이를 코드에 import시키는 일이다.

```python
import numpy as np
from sklearn import linear_model # scikit-learn 모듈을 가져온다

regr = linear_model.LinearRegression()
```

선형회귀를 위한 입력 데이터 x를 2차원 리스트에 넣는다. 다음으로 y변수를 초기화한다. 이 데이터를 이용하여 선형회귀 학습을 시작한다.

```python
x = [[164], [179], [162], [170]] # 다중회귀에도 사용하도록 함
y = [53, 63, 55, 59] # y = f(x)의 결과
regr.fit(x, y)
```

주의할 점은 학습 데이터는 여러 변수에 대해서 다중회귀분석을 실시하기 위해서 반드시 2차원 배열이어야 한다는 점이다.

선형 회귀 분석을 적용하려면 fit() 함수에 x와 y를 전달한다. 점들의 분포를 표현하는 직선이 세 종류가 있는데 sklearn의 LinearRegressin 객체는 가능한 직선들 가운데 가장 데이터를 잘 따르는 직선을 계산해낼 수 있다.

## 14.7 선형 회귀 학습결과를 확인하고 예측하기

선형회귀 방정식을 구하는 기계 학습은 sklearn에서 간단히 이루어져있다.

```python
regr.fit(x, y)
```

학습된 결과를 확인하고 이를 이용하여 데이터 예측을 수행한다.

```python
coef = regr.coef_ # 직선의 기울기
intercept = regr.intercept_ # 직선의 절편
score = regr.score(x, y) # 학습된 직선이 데이터를 얼마나 잘 따르나

print("y =", coef, "* x +", intercept)
print("The score of this line for the data :", score)
```

학습한 데이터를 바탕으로 새로운 입력이 주어지면 그에 맞는 결과를 예측할 수 있다. 이 예측은 학습을 통해 찾아 놓은 직선의 방정식을 통해 이루어질 것이다.

값을 예측하는 함수는 predict()이다. 이 함수는 매개변수는 예측에 필요한 입력 데이터이고, 이 데이터는 선형회귀 학습을 할 때 사용한 입력 x와 같이 2차원 배열이다.

```python
input_data = [[180], [185]]
```

이것을 predict() 함수에 넘겨주면 예측결과를 얻을 수 있다.

```python
result = regr.predict(input_data)
print(result)
```

## 14.8 선형회귀로 예측하기 : 키와 몸무게는 상관관계가 있을까

```python
regr.predict([[169]])
```

선형회귀를 그래프로 그리기 위해 matplotlib 라이브러리를 사용한다.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model # scikit-learn 모듈을 가져온다

regr = linear_model.LinearRegression()

x = [[164], [179], [162], [170]] # 선형회귀의 입력은 2차원으로 만들어야 함
y = [53, 63, 55, 59] # y = f(x)의 결과값
regr.fit(x, y)

# 학습 데이터와 y값을 산포도로 그린다
plt.scatter(x, y, color='black')

# 학습 데이터를 입력으로 하여 예측값을 계산한다
y_pred = regr.predict(x)

# 학습 데이터와 예측값으로 선그래프를 그린다
# 계산된 기울기와 y 절편을 가지는 직선이 그려진다
plt.plot(x, y_pred, color='blue', linewidth=3)
plt.show()
```

x, y 값을 화면에 출력한 후 선형회귀 모델을 적용하여 직선을 그렸다. 선형회귀 모델이 적용된 직선이 데이터의 분포를 효과적으로 표현하는 것을 볼 수 있다.

학습을 마치고 얻는 계수와 절편을 이용하여 테스트를 해 볼 데이터가 존재하지 않는다. 그래서 원래의 데이터 중에서 일부는 학습에 사용하고, 남은 데이터를 가지고 학습결과가 예측을 잘 하는지 테스트해 보는 것이 좋다. 실제로 많은 기계학습에서 이러한 전략을 사용하여 학습의 수준을 평가한다.

```python
# 학습 데이터와 테스트 데이터를 분리한다
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.2)
```

x_train과 y_train과 같은 학습용 데이터를 이용하여 선형 회귀 모델로 학습을 수행한다.

```python
# 학습 데이터와 테스트 데이터를 분리한다
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(diabetes.data[:, np.newaxis, 2], diabetes.target, test_size=0.2)

regr = LinearRegression()
regr.fit(x_train, y_train)
```

전체 데이터를 학습시킨 것과 거의 유사한 계수와 절편을 얻은 것을 확인할 수 있다. 이것은 데이터의 80 퍼센트를 이용하여 학습한 결과이다. 이것을 이용하여 테스트용 데이터를 학습을 통해 얻은 선형 함수가 얼마나 잘 따르는지를 확인하기 위해 score() 함수를 사용할 수 있다.

```python
score = regr.score(x_train, y_train)
print(score)

score = regr.score(x_test, y_test)
print(score)
```

선형회귀 학습이 마치고 학습용 데이터는 찾아낸 선형함수를 따르는 정도가 0.349 정도인데, 테스트용 데이터는 0.319 정도가 나오고 있다. 테스트용 데이터는 학습에 사용되지 않았으니 자연스러운 결과이다.

## 14.15 K-NN 알고리즘을 이용한 분류

K-NN 방법은 이전의 데이터를 바탕으로 새로운 데이터를 가장 가까운 것으로 분류하는 방법이다. 이 분류 방법은 가장 가까운 이웃에만 의존하기 때문에 최근접 이웃 방법이라고 불린다.

하지만 가장 가까운 이웃 하나만 보고 클래스를 판단하는 것은 문제가 있을 수 있다. 대신에 우리는 가장 가까운 k개의 도형을 확인하여 그들 중 다수인 쪽으로 새로운 데이터를 속하게 한다.

K-NN 알고리즘은 K의 값에 따라 결과가 달라진다. 판단의 편의를 위해 K를 홀수로 취하는 것이 좋다.

새 데이터와 가까운 도형들은 더 높은 가중치를 얻고, 먼 도형들은 더 낮은 가중치를 가지도록 한다. 그런 다음 모든 도형의 가중치를 전부 합한다. 그리고 보다 더 큰 가중치를 얻은 도형이 새로운 데이터의 클래스가 된다. 이것을 수정된 K-NN이라고 한다.

K-NN 방법은 특정 공간에 있는 모든 데이터에 대한 정보가 필요하다. 가장 가까운 이웃을 찾기 위해 새로운 데이터에서 모든 기존 데이터까지의 거리를 확인해야하기 때문이다. 따라서 데이터와 클래스가 많이 있다면 많은 메모리 공간과 계산 시간이 필요하다는 단점이 있다.

## 14.18 K-NN 알고리즘을 적용해보자

train_test_aplit() 함수를 이용하여 전체 데이터의 80%를 학습용 데이터로 사용하여 이 모델이 나머지 20%의 테스트 데이터를 잘 예측하는지 확인해 볼 수 있다.

```python
# (80:20)으로 분할한다
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

KNeighborClassifier를 사용하여 학습과 테스트를 진행해 보자. 그리고 sklearn의 metrics를 사용하여 accuracy_score() 함수를 불러올 것이며 이 함수를 이용해서 K-NN의 예측치 y_pred와 실제값 y_test와의 차이를 확인해 볼 것이다.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

num_neigh = 1
knn = KNeighborsClassifier(n_neighbors = num_neigh)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
scores = metrics.accuracy_score(y_test, y_pred)
print('n_neighbors가 {0:d}일때 정확도 : {1:3f}'.format(num_neigh, scores))
```

k의 최적 값을 선택하는 것은 매우 중요한데, 일반적으로 k의 값이 증가함에 따라 정확도가 증가하지만 너무 커질 경우 정확도는 다시 감소하는 경향이 있다.

기대수명에 영향을 주는 요소들이 무엇인지 궁금하여 데이터를 이용하여 쌍 그림을 그려 보았다. 쌍 그림은 주어진 데이터의 열들이 서로 어떠한 상관관계가 있는지 산포도 그래프로 보여준다.

```python
sns.pairplot(life[['Life expectancy', 'Alcohol', 'Percentage expenditure', 'Measles', 'Polio', 'BMI', 'GDP', 'Thinness 1-19 years']])
plt.show()
```

그림을 살펴보면 BMI와 GDP 같은 데이터가 상당한 연관도를 보이고 있는 것을 확인할 수 있다.

## 14.25 간단한 회귀모델을 만들자

사망률에 크게 영향을 미치지 않는 특징들을 모두 포함하면 학습이 잘 되지 않거나, 예측의 신뢰도가 오히려 떨어질 수 있다.

```python
x = life[['Alcohol', 'Percentage expenditure', 'Polio', 'BMI', 'GDP', 'Thinness 1-19 years']]
y = life['Life expectancy']
```

그리고 train_test_split() 함수를 이용하여 80%의 학습 데이터와 20%의 테스트 데이터로 나눈다.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

사이킷런의 LinearRegression 모델을 사용하여 학습을 진행해 보자.

```python
from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
lin_model.fit(x_train, y_train)
y_test_predict = lin_model.predict(x_test)
```

평균제곱오차인 MSE는 오차의 제곱값을 평균하였으므로 이 값에 제곱근을 취하여 평균오차값에 가깝도록 보정한 값이 평균제곱근오차이다.

```python
from sklearn.metrics import mean_squared_error
y_test_predict = lin_model.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, y_test_predict))
print('RMSE =', rmse)
```

정답 레이블 y_test와 예측하여 얻은 값 y_test_predict를 순서쌍으로하는 데이터의 산포도를 그려보면 예측을 얼마나 잘 하는지 살펴볼 수 있다.

```python
plt.scatter(y_test, y_test_predict)
plt.show()
```

## 14.26 과적합과 과소적합

기계학습에서 흔히 볼 수 있는 가능성은 과적합이다. 과적합이란 학습하는 데이터에서는 성능이 뛰어나지만 새로운 데이터에 대해 적용했을 때, 성능이 잘 나오지 않는 모델을 만들어내는 일이다. 과적합이 나오는 이유는 데이터에 섞인 잡음까지 학습하기 때문이다. 이와 반대로 데이터에 대한 학습이 지나치게 대충 이루어져 학습 데이터와 다른 데이터 모두에 대해 예측을 제대로 하지 못하는 모델을 만드는 일을 과소적합이라고 한다.

데이터를 설명하는 곡선을 만들기 위하여 아주 높은 지수의 다항식으로 학습시키면 모든 학습 데이터를 정확하게 학습할 것이다. 하지만 새로운 데이터를 제시한다면, 아마도 잘못된 출력이 나올 가능성이 매우 높다.

- 기계학습은 인공 지능의 한 분야로 컴퓨터가 데이터를 이용하여 예측 성능을 갖게 하는 것이다
- 기계학습의 큰 분야는 지도학습, 자율학습, 그리고 강화학습이 있다
- 선형 회귀 분석은 데이터의 분포와 가장 잘 일치하는 선형 방정식을 찾는 것이다
- 사이킷런 모듈은 회귀 모형에 기반한 기계학습을 쉽게 구현할 수 있도록 해 준다
- 데이터를 분류하는 알고리즘의 대표적인 예가 K-NN 알고리즘이다
- 기계학습을 통해 만들어진 모델은 과적합이나 과소적합을 보일 수 있다
- 학습이 적절히 이루어지게 만드려면 과적합시에는 데이터를 늘이고, 과소적합에는 특징을 늘려볼 수 있다

## 12.1 엑셀보다 빠른 일처리는 판다스로

엑셀은 행과 열로 이루어진 표에 입력된 데이터를 처리하는데 탁월한 성능을 보이고 있다.  2차원 형태의 데이터는 개발자가 편리하게 각 요소나 행, 열에 접근할 수 있다. 넘파이가 2차원 행렬 형태의 데이터를 지원하지만, 넘파이는 데이터의 속성을 표시하는 행이나 열의 레이블을 가지고 있지 않다는 한계가 있다. 하지만 파이썬의 판다스 패키지를 사용하면 이러한 문제를 해결할 수 있다.

판다스는 넘파이를 기반으로 하기 때문에 처리속도가 매우 빠르며, 데이터프레임은 행과 열로 구조화되어 있고, 다양한 함수를 지원한다. 행과 열에는 레이블이 붙어 있어서 매우 편리하게 각 행과 열에 접근할 수 있으며, 데이터프레임을 결합하거나 그 형태를 변경하는 일도 간편하게 할 수 있다.

< 판다스의 특징 >
1. 빠르고 효율적이며 다양한 표현력을 갖춘 자료 구조
2. 다양한 형태의 데이터에 적합 : 이중 자료형의 열을 가진 테이블 데이터, 시계열 데이터, 다양한 관측 통계 데이터
3. 핵심구조
   - 시리즈 : 1차원 구조를 가진 하나의 열
   - 데이터프레임 : 복수의 열을 가진 2차원 데이터
4. 판다스가 잘 하는 일 : 결측 데이터 철, 데이터 추가 삭제, 데이터 정렬과 다양한 데이터 조작

## 12.3 CSV 라고 들어봤니

CSV는 테이블 형식의 데이터를 저장하고 이동하는 데 사용되는 구조화된 텍스트 파일 형식이다. CSV는 쉼표로 구분한 변수의 약자이다. 따라서 기본적으로 쉼표를 이용하여 데이터를 구분한다. 탭이나 콜론, 세미콜론 등의 구분자도 사용할 수 있다.

CSV 파일은 필드를 나타내는 열과 레코드를 나타내는 행으로 구성된다. 파일의 첫 번쨰 레코드에는 열 제목이 포함되어 있을 수 있다. 파일의 크기를 알 수 없고 잠재적으로 크기가 큰 경우 한 번에 모든 레코드를 읽지 않는 것이 좋다. 이때는 특정한 크기만큼의 데이터를 읽어서 처리한 뒤에, 다음 데이터를 가져오는 방식으로 할 수 있다.

## 12.4 CSV 데이터의 내용을 읽어 보자

파이썬 모듈 csv는 csv reader와 csv writer를 제공한다. 두 객체 모두 파일 핸들을 첫 번쨰 매개 변수로 사용한다.

## 12.5 CSV에서 원하는 데이터를 뽑아 보자

다운 받은 데이터에서 원하는 데이터만 추출하여 사용하고 싶으면 row[]를 사용하여 해당 데이터의 인덱스로 찾을 수 있다.

```python
import csv

f = open('C:\kdt\weather.csv') # csv 파일을 열어서 f에 저장
data = csv.reader(f) # csv의 reader() 함수를 이용하여 읽는다
header = next(data) # 헤더를 제거
for row in data:
    print(row[3], end=',') # 평균풍속만 출력하고, 쉼표로 연결한다
f.close() # 파일을 닫는다
```

문자열을 실수로 변환하는 함수는 float()이다.

```python
import csv

f = open('C:\kdt\weather.csv') # csv 파일을 열어서 f에 저장
data = csv.reader(f) # csv의 reader() 함수를 이용하여 읽는다
header = next(data) # 헤더를 제거
max_wind = 0

for row in data: # 반복 루프를 사용하여 데이터를 읽는다
    if row[2] == "": # 평균 풍속 데이터가 없는 경우 0을 처리
        wind = 0
    else:
        wind = float(row[2]) # 평균 풍속 데이터를 실수로 변환해 저장
    if max_wind < wind: # 최대 풍속을 갱신하는지 검사
        max_wind = wind # 현재까지의 최대 풍속보다 크면 새로 기록

print("지난 10년간 울릉도의 최대 풍속은", max_wind, 'm/s')
```

## LAB 12-1 울릉도는 몇 월에 바람이 가장 강할까?

```python
import csv
import matplotlib.pyplot as plt

f = open('C:\kdt\weather.csv') # csv 파일을 열어 f에 저장
data = csv.reader(f) # reader() 함수로 읽기
header = next(data) # 헤더를 제거

monthly_wind = [0 for x in range(12)] # 매달 풍속을 담을 리스트, 초기화 0
days_counted = [0 for x in range(12)] # 각 달마다 측정된 일수, 초기화 0

for row in data:
    month = int(row[0][5:7]) # 0번 열에서 달 정보 추출
    if row[3] != '': # 풍속 데이터 존재하는지 확인
        wind = float(row[3]) # 풍속을 얻어 온다
        monthly_wind[month-1] += wind # 해당 달에 풍속 데이터 추가
        days_counted[month-1] += 1 # 해당 달의 일수를 증가
        
for i in range(12):
    monthly_wind[i] /= days_counted[i] # 일수로 나누어 월평균 구하기

plt.plot(monthly_wind, 'blue')
plt.show()

f.close() # 파일을 닫는다
```

## 12.6 판다스의 데이터 구조 : 시리즈와 데이터 프레임

csv 모듈 이외에도 csv 데이터를 처리할 수 있는 모듈이 있다.
- 시리즈(1차원) : 레이블이 붙어있는 1차원 벡터
- 데이터프레임(2차원) : 행과 열로 되어있는 2차원 테이블, 각 열은 시리즈로 되어 있다

데이터 구조는 모두 넘파이 배열을 이용하여 구현된다. 따라서 속도가 빠르다. 모든 데이터 구조는 값을 변경할 수 있으면, 시리즈를 제외하고는 크기도 변경할 수 있다. 각 행과 열은 이름이 부여되며, 행의 이름을 인덱스, 열의 이름을 컬럼스라 부른다.

시리즈는 동일 유형의 데이터를 저장하는 1차원 배열이다. 판다스의 시리즈 데이터를 만드는 것은 seried 클래스를 이용한다.

```python
import numpy as np
import pandas as pd
series = pd.Series([1, 3, 4, np.nan, 6, 8])
series
```

1차원 벡터를 생성하였다. 이것이 판다스에서 사용하는 가장 기초적인 데이터 구조인 시리스이다.

데이터프레임은 시리즈 데이터가 여러 개 모여서 2차원적 구조를 갖는 것이다. 이 데이터프레임은 판다스가 데이터를 분석할 떄 사용하는 기본적 틀이다. 하나의 데이터프레임은 행과 열로 구분할 수 있는데, 하나의 행은 여러 종류의 데이터를 담고 있다. 각 열은 동일한 자료형을 가진 시리즈이다.

```python
name_series = pd.Series(['김수안', '김수정', '박동윤', '강이안', '강지안'])
age_series = pd.Series([19, 23, 22, 19, 16])
sex_series = pd.Series(['여', '여', '남', '여', '남'])
grade_series = pd.Series([4.35, 4.23, 4.25, 4.37, 4.25])
print(name_series, age_series, sex_series, grade_series)
```

시리즈들을 모아 하나의 데이터프레임을 만들 수 있다. 이 때는 판다스의 DataFrame 클래스를 사용한다. DataFrame을 생성하기 위한 인자로 시리즈들을 나열하면 되는데, 딕셔너리 구조로 입력을 한다. 딕셔너리의 키는 시리즈가 차지할 열의 이름, 값은 시리즈가 된다.

```python
df = pd.DataFrame({'이름': name_series, '나이': age_series, '성별': sex_series, '평점': grade_series})
print(df)
```

## 12.13 데이터를 간편하게 분석할 수 있는 기능이 있다

외부 파일을 읽어서 데이터프레임을 생성한 후 필요한 행과 열을 선택할 수 있다. 데이터 프레임이 저장한 데이터를 간단히 분석하려면 describe() 함수를 호출해주면 된다.

```python
import pandas as pd
weather = pd.read_csv('c:\kdt\weather.csv', index_col=0, encoding='cp949')

print(weather.describe())
```

평균값(mean), 표준편차(std), 최솟값(min), 최댓값(max) 등을 쉽게 알 수 있다.

```python
print('평균 분석 ---------------------')
print(weather.mean())
print('표준편차 분석 ---------------------')
print(weather.std())
```

판다스 표준편차와 넘파이 표준편차를 구하는 방법은 다르다. 계산한 결과도 다른 값이 나온다.

why? 판다스의 표준편차는 베셀 보정을 적용하는데, 표준편차를 구할 때 표본 크기 n 대신에 n-1을 적용하는 것으로 모분산 추정에서 편향을 보정하는 역할을 한다.

## 12.19 데이터 구조를 변경해 보자

데이터프레임을 csv를 읽어서 생성할 수도 있지만, 딕셔너리 데이터를 이용하여 생성할 수도 있다. 이때 키는 열의 레이블이 되고, 딕셔너리의 키에 딸린 값은 열을 채우는 데이터를 가진 리스트가 된다.

```python
import pandas as pd

df_1 = pd.DataFrame({'item' : ['ring0', 'rinf0', 'ring1', 'ring1'],
                     'type' : ['Gold', 'Silver', 'Gold', 'Bronze'],
                     'price' : [20000, 10000, 50000, 30000]})
print(df_1)
```

## 12.20 concat() 함수로 데이터프레임을 합쳐보자

일반적으로 데이터들은 하나의 큰 테이블로 저장되지 않고, 작은 테이블로 나누어져 있는 경우가 많다. 데이터를 하나로 합치는 방법 가운데 하나인 concat()를 사용할 수 있다.

```python
df_1 = pd.DataFrame({'A' : ['a10', 'a11', 'a12'],
                     'B' : ['b10', 'b11', 'b12'],
                     'C' : ['c10', 'c11', 'c12']}, index = ['가', '나', '다'])
df_2 = pd.DataFrame({'B' : ['b23', 'b24', 'b25'],
                     'C' : ['c23', 'c24', 'c25'],
                     'D' : ['d23', 'd24', 'd25']}, index = ['다', '라', '마'])
```

두 데이터프레임을 결합하기 위해서는 concat() 함수를 사용하면 된다. 첫 인자로는 모아서 합칠 데이터프레임들을 리스트로 넘겨준다.

```python
df_3 = pd.concat([df_1, df_2]) # df_1, df_2 두 데이터프레임을 합쳐서 df_3을 생성
print(df_3)
```

디폴트 축이 0일 경우 행을 늘려 데이터 테이블을 붙여 나간다. 데이터를 찾을 수 없는 경우에는 결손값으로 처리된다.

## LAB 12-4 다양한 방법으로 concat 적용해 보기

```python
import pandas as pd

df_1 = pd.DataFrame({'A' : ['a10', 'a11', 'a12'],
                     'B' : ['b10', 'b11', 'b12'],
                     'C' : ['c10', 'c11', 'c12']}, index = ['가', '나', '다'])
df_2 = pd.DataFrame({'B' : ['b23', 'b24', 'b25'],
                     'C' : ['c23', 'c24', 'c25'],
                     'D' : ['d23', 'd24', 'd25']}, index = ['다', '라', '마'])

print(pd.concat([df_1, df_2], axis=0, join='outer'), '\n')
print(pd.concat([df_1, df_2], axis=0, join='inner'), '\n')
print(pd.concat([df_1, df_2], axis=1, join='outer'), '\n')
print(pd.concat([df_1, df_2], axis=1, join='inner'))
```

## 12.21 데이터베이스 join 방식의 데이터 병합 - merge

데이터베이스는 조인이라는 연산을 지원한다. 조인 연산과 같은 방식의 데이터 병합을 지원하는 판다스 함수가 merge() 함수이다.

df_1.merge(df_2, 다양한 선택 사항을 결정하는 키워드 인자들..)

how에 'outer'를 넣으면 키 값들의 합집합을 기준으로 테이블을 만들 수 있다.

```python
df_3 = df_1.merge(df_2, how='outer', on='B')
print(df_3)
```

## 12.22 인덱스를 키로 활용하여 merge 적용해 보기

두 데이터를 결합할 때 사용할 키가 될 레이블을 지정하면, 해당 레이블에 있는 값들을 이용하여 테이블을 생성한다. 그런데 원래 테이블에 있던 인덱스는 사라진 것을 확인할 수 있다.

많은 경우 인덱스가 키의 역할을 수행하는 경우도 있다. 이러한 경우 인덱스를 키로 사용하라고 할 수도 있다.

```python
df_1.merge(df_2, how='outer', left_index=True, right_index=True)
```

인덱스가 병합의 키가 되도록 하는 방식을 아래와 같다.

```python
df_3 = df_1.merge(df_2, how='outer', left_index=True, right_index=True)
print(df_3)
```

## LAB 12-5 다양항 방법으로 merge 적용해 보기

```python
import pandas as pd

df_1 = pd.DataFrame({'A' : ['a10', 'a11', 'a12'],
                     'B' : ['b10', 'b11', 'b12'],
                     'C' : ['c10', 'c11', 'c12']}, index = ['가', '나', '다'])
df_2 = pd.DataFrame({'B' : ['b23', 'b24', 'b25'],
                     'C' : ['c23', 'c24', 'c25'],
                     'D' : ['d23', 'd24', 'd25']}, index = ['다', '라', '마'])

print('left outer \n', df_1.merge(df_2, how='left', on='B'), '\n')
print('right outer \n', df_1.merge(df_2, how='right', on='B'), '\n')
print('full outer \n', df_1.merge(df_2, how='outer', on='B'), '\n')
print('inner \n', df_1.merge(df_2, how='inner', on='B'))
```

## 12.23 데이터를 크기에 따라 나열하자 : 정렬

데이터 분석에 있어서 데이터를 크기에 따라 순서대로 나열하는 정렬은 필수적이다. sort_values() 함수를 호출하여 사용할 수 있다.

```python
import pandas as pd
import matplotlib.pyplot as plt

countries_df = pd.read_csv('c:\kdt\countries.csv', index_col=0)
sorted = countries_df.sort_values('population')
print(sorted)
```

새로운 변수를 도입하지 않고 원래의 데이터프레임 자체가 정렬된 상태로 바뀌게 하고 싶으면 inplace 키워드 매개변수를 사용한다.
